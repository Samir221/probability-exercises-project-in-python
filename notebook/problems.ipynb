## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import random

def roll_dice():
    return random.randint(1, 6), random.randint(1, 6)

def is_even_or_greater_than_seven(total):
    return total % 2 == 0 or total > 7

def simulate_dice_rolls(num_simulations):
    success_count = 0
    
    for _ in range(num_simulations):
        dice1, dice2 = roll_dice()
        total = dice1 + dice2
        
        if is_even_or_greater_than_seven(total):
            success_count += 1
            
    return success_count / num_simulations

# Run the simulation 1000 times
num_simulations = 1000
probability = simulate_dice_rolls(num_simulations)

print(f"Estimated Probability: {probability}")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
import random

# Create the box of balls
ball_box = {}

for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i >= 10) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

# Function to draw 5 balls with replacement
def draw_balls(ball_box):
    drawn_balls = [ball_box[random.randint(0, 59)] for _ in range(5)]
    return drawn_balls

# Function to calculate the probability
def simulate_draws(num_simulations):
    count_3_white_2_red = 0
    count_all_same_color = 0
    
    for _ in range(num_simulations):
        drawn_balls = draw_balls(ball_box)
        
        white_count = drawn_balls.count("White")
        red_count = drawn_balls.count("Red")
        green_count = drawn_balls.count("Green")
        
        # Condition for 3 white and 2 red
        if white_count == 3 and red_count == 2:
            count_3_white_2_red += 1
            
        # Condition for all same color
        if white_count == 5 or red_count == 5 or green_count == 5:
            count_all_same_color += 1
    
    probability_3_white_2_red = count_3_white_2_red / num_simulations
    probability_all_same_color = count_all_same_color / num_simulations
    
    return probability_3_white_2_red, probability_all_same_color

# Run the simulation 1000 times
num_simulations = 1000
prob_3_white_2_red, prob_all_same_color = simulate_draws(num_simulations)

print(f"Estimated Probability of 3 White and 2 Red: {prob_3_white_2_red}")
print(f"Estimated Probability of All Same Color: {prob_all_same_color}")
```
